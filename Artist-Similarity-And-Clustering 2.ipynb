```python
song_features = acoustic_features.merge(lyrics, on = 'song_id')
song_features
```

```python
df1 = song_features.merge(songs, on = 'song_id', suffixes=('', '_songs'))
len(df1)
```

```python
df2 = df1.merge(tracks, on = 'song_id', suffixes=('', '_tracks'))
len(df2)
```

```python
df3 = df2.merge(albums, on = 'album_id', suffixes=('', '_albums'))
len(df3)
```

```python
df4 = df3.merge(releases, on = 'album_id', suffixes=('', '_releases'))
len(df4)
```

```python
df5 = df4.merge(artists, on = 'artist_id', suffixes=('', '_artists'))
len(df5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# basic statistics
display(df5.describe())

# Plotting distributions of numerical features
df5.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Correlation matrix
corr_matrix = df5.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
df5.head(10)
```

```python

# Grouping by artist_id and calculating the mean popularity for each artist
artist_popularity = df5.groupby('artist_id')['popularity'].mean().reset_index()

# Merging artist popularity with the artists df for artist names
artist_popularity = artist_popularity.merge(artists[['artist_id', 'name']], on='artist_id')

# Sort artists by popularity
artist_popularity = artist_popularity.sort_values(by='popularity', ascending=False)

# Displaying the top 10 artists with similar popularity
artist_popularity.head(15)
```

```python
# Changing the datatype of the 'followers' column from str to int
df5['followers'] = df5['followers'].astype(int)
```

```python
# Grouping by artist_id and calculating the mean followers for each artist
artist_followers = df5.groupby('artist_id')['followers'].mean().reset_index()

# Merging artist followers with the artists df for artist names
artist_followers = artist_followers.merge(artists[['artist_id', 'name']], on='artist_id')

# Sort artists by followers
artist_followers = artist_followers.sort_values(by='followers', ascending=False)

# Displaying the top 10 artists with similar followers
artist_followers.head(15)
```

```python
# Calculate the correlation between popularity and followers
correlation = df5['popularity'].corr(df5['followers'])

# Display the correlation
correlation

# Plotting the correlation
plt.figure(figsize=(10, 6))
sns.scatterplot(x='popularity', y='followers', data=df5)
plt.title(f'Correlation between Popularity and Followers: {correlation:.2f}')
plt.xlabel('Popularity')
plt.ylabel('Followers')
plt.show()
```

```python
# Perform a linear regression to check for a trend between popularity and followers
from sklearn.linear_model import LinearRegression
import numpy as np

# Reshape the data for sklearn
X = df5['popularity'].values.reshape(-1, 1)
y = df5['followers'].values

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Get the slope (coefficient) and intercept of the line
slope = model.coef_[0]
intercept = model.intercept_

# Plot the data and the trend line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='popularity', y='followers', data=df5)
plt.plot(df5['popularity'], model.predict(X), color='red', linewidth=2)
plt.title(f'Trend between Popularity and Followers: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Popularity')
plt.ylabel('Followers')
plt.show()
```
